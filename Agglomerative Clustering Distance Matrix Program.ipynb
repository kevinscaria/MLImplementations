```python
def ClusterManual(k1):
    ix=k1.index
    l,x=[],[]
    for i in range(len(k1)):  #Creating A List from the distance matrix
        for j in range(len(k1)):
            if k1.iloc[i,j]!=0:
                l.append([[i,j],k1.iloc[i,j]])
                x.append(k1.iloc[i,j])
    
    #Finding the smallest distance
    np.min(x)
    g=[]
    for i in range(len(l)):
        if l[i][1]==np.min(x):
            g.append(l[i][0])  
    
    #Recalculating distance from the newly clustered points (Smallest distant points)
    x1=[]
    for i in range(len(k1)):
        for j in range(len(k1)):
            if i==j:
                x1.append(0)
            elif j==g[0][0]:
                x1.append(max(k1.iloc[g[0][1],i],k1.iloc[g[0][0],i]))
            elif i==g[0][0]:
                x1.append(max(k1.iloc[j,g[0][0]],k1.iloc[j,g[0][1]]))
            else:
                x1.append(k1.iloc[i,j])

    k2=pd.DataFrame(np.array(x1).reshape(len(k1),-1))
    
    #COLUMN and INDEX naming
    k2.columns=[(ix[g[0][0]],ix[g[0][1]]) if i == g[0][0] else ix[i] for i in range(len(ix))]
    k2.index=[(ix[g[0][0]],ix[g[0][1]]) if i == g[0][0] else ix[i] for i in range(len(ix))]
    k2.drop(ix[g[0][1]],axis=1,inplace=True)
    k2.drop(ix[g[0][1]],axis=0,inplace=True)
    return k2

#Main Function
def ClusterNest(df):
    from IPython.display import display
    from scipy.spatial.distance import cdist
    k=pd.DataFrame(cdist(df,df,'euclidean')).round(2)
    print('\n**** First Distance Matrix ****')
    df1=pd.DataFrame(cdist(df,df)).round(2)
    df1.index=df.index
    df1.columns=df.index
    display(df1)
    lc=len(df1.columns)
    xx=1
    while lc > 1:
        y=ClusterManual(df1)
        lc=lc-1
        df1=y
        print('\n ***** Updated Distance Matrix {} ****'.format(xx))
        display(df1)
        xx=xx+1
```
